```python
%matplotlib inline
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(-1, 1, 201)
x = 0.18*np.cos(2*np.pi*3.85*t)
xn = x + np.random.randn(len(t)) * 0.08
```

```python
plt.figure(figsize=(20,5))
plt.plot(xn)
```

```python
# Create an order 3 lowpass butterworth filter:
b, a = signal.butter(3, 1/4)
y = signal.filtfilt(b, a, xn)

# Plot the filtered data
plt.figure(figsize=(20,5))
plt.plot(y, color="b")

plt.plot(xn, color="r")
```

```python
# Create an order 3 lowpass butterworth filter:

# Plot the filtered data
plt.figure(figsize=(20,5))
plt.plot(xn, color="b")

legends = []
for Fc in range(10,20,2):
    b, a = signal.butter(3, 1/Fc)
    y = signal.filtfilt(b, a, xn)
    plt.plot(y)
    legends.append("Fc={}".format(Fc))

plt.legend(legends)
```

```python
# High frequency terms after applying a butterworth filter on xn vs not applying a filter

from scipy.fftpack import fft, ifft
yf1 = np.abs(fft(xn))
plt.plot(yf1[0:len(yf1)//2], color="b")
yf2 = np.abs(fft(y))
plt.plot(yf2[0:len(yf2)//2], color="r")
plt.legend(["Not Filtered","Filtered"])
```
