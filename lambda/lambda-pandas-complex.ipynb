```python
import pandas as pd

fifa = pd.read_csv('fifa.csv')

fifa.head()
```

```python
fifa.dtypes
```

```python
def age_dog(x): 
    return x*7

fifa['dog'] = fifa.Age.apply(age_dog)
fifa.head()
```

```python
fifa['dog'] = fifa.Age.apply(lambda x: age_dog(x))
fifa.head()
```

```python
def cat_age(age, over): 
    return age*7 + over

fifa['cat'] = fifa.apply(lambda row: cat_age(row['Age'], row['Overall']), axis=1)
fifa.head()
```
