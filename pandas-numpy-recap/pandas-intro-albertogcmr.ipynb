```python
def mensaje(name, age, survived): 
    if survived == 1 or survived: 
        survived = 'sobrevivi√≥'
    else: 
        survived = 'falleci√≥'
    try: 
        age = int(age)
    except: 
        age = 'edad desconocida'

    return "{} de {} a√±os {}.".format(name, age, survived)
```

```python
# df['newcolumn'] = df.apply(lambda x: fxy(x['A'], x['B']), axis=1)

# https://stackoverflow.com/questions/19914937/applying-function-with-multiple-arguments-to-create-a-new-pandas-column/19976286

df['message'] = df.apply(lambda x: mensaje(x['Name'], x['Age'], x['Survived']), axis=1)
df.head()
```
